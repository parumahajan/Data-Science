```python
import pandas as pd
from google.colab import files
import io

# Upload file
uploaded = files.upload()

# Load the dataset
file_name = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# 1. Explore data using DataFrames
print("First 5 rows of the dataset:")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# 2. Index objects
print("\nIndex of DataFrame:")
print(df.index)

# 3. Re-indexing
df_reindexed = df.reindex(sorted(df.columns), axis=1)
print("\nData after reindexing columns alphabetically:")
print(df_reindexed.head())

# 4. Drop an entry (Dropping the first row as an example)
df_dropped = df.drop(index=0)
print("\nData after dropping first row:")
print(df_dropped.head())

# 5. Selecting Entries (Selecting specific columns as an example)
selected_entries = df.iloc[:5, :3]  # First 5 rows and first 3 columns
print("\nSelected Entries:")
print(selected_entries)

# 6. Data Alignment
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([4, 5, 6], index=['b', 'c', 'd'])
aligned = s1 + s2
print("\nData Alignment Example:")
print(aligned)

# 7. Rank and Sort
print("\nRanking Values in a column:")
print(df.rank())
print("\nSorting Data by a column (assuming numeric column):")
print(df.sort_values(by=df.columns[0], ascending=True).head())

# 8. Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# 9. Index Hierarchy (Creating a MultiIndex example)
multi_index = pd.MultiIndex.from_tuples([(1, 'A'), (1, 'B'), (2, 'A'), (2, 'B')], names=['Index1', 'Index2'])
data = [[10, 20], [30, 40], [50, 60], [70, 80]]
df_multi = pd.DataFrame(data, index=multi_index, columns=['Column1', 'Column2'])
print("\nMultiIndex DataFrame:")
print(df_multi)
```
